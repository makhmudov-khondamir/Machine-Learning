### **Project Title: Analyzing and Predicting Student Performance Based on Computer Resources**

**Objective:**

This project aims to analyze and predict how varying numbers of computers impact student performance. Two different models are used to provide comprehensive insights:

1. **Logistic Model Analysis:**
   - **Determine the Saturation Point:** Identify the optimal number of computers where student performance benefits level off. The logistic growth model is used to pinpoint the saturation point, beyond which additional computers yield diminishing returns.
   - **Predict Future Performance:** Use the logistic model to forecast student performance for future numbers of computers, focusing on the performance levels achievable with different amounts of resources.

2. **Linear Regression Analysis:**
   - **Analyze Incremental Impact:** Assess how each additional computer influences student performance through a linear relationship. The linear regression model quantifies the incremental benefit of each computer.
   - **Predict Future Performance:** Use the linear regression model to estimate student performance for various future scenarios with increased numbers of computers, providing a clear understanding of performance improvements.

**Code Overview:**

- **Logistic Model Analysis:** The project fits a logistic growth model to historical data on the number of computers and student performance. The model identifies the saturation point and predicts performance for future numbers of computers. The code involves data preparation, model fitting, prediction, and result interpretation.
  
- **Linear Regression Analysis:** The project applies linear regression to the same data to understand the relationship between the number of computers and student performance. It estimates how performance changes with each additional computer and makes predictions for future scenarios. The code includes data preparation, model fitting, prediction, and result interpretation.


###**Saturation Level Analysis**


```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Sample data: Replace with your actual data
data = pd.DataFrame({
    'number_of_computers': [100, 200, 300, 400, 500, 600, 700, 800],
    'student_performance': [10, 20, 30, 40, 45, 47, 48, 48]
})

# Define the logistic growth model
def logistic(x, L, k, x0):
    return L / (1 + np.exp(-k * (x - x0)))

# Extract data for fitting
x_data = data['number_of_computers']
y_data = data['student_performance']

# Fit the logistic model to the data
params, covariance = curve_fit(logistic, x_data, y_data, p0=[max(y_data), 1, np.median(x_data)])

# Generate fitted values
x_fit = np.linspace(min(x_data), max(x_data), 100)
y_fit = logistic(x_fit, *params)

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(x_data, y_data, label='Observed Data', color='blue')
plt.plot(x_fit, y_fit, label='Fitted Logistic Curve', color='red')
plt.xlabel('Number of Computers')
plt.ylabel('Student Performance')
plt.title('Saturation Level Analysis')
plt.legend()
plt.grid(True)
plt.show()

# Output the parameters
L, k, x0 = params
print(f"Fitted Logistic Model Parameters:")
print(f"L (Maximum Benefit): {L}")
print(f"k (Growth Rate): {k}")
print(f"x0 (Saturation Point): {x0}")

# Interpretation
print("\nInterpretation:")
print(f"The saturation point, x0, indicates the number of computers where additional resources provide diminishing returns.")
print(f"Benefit levels off around {x0} computers. Additional computers beyond this point yield minimal improvement.\n")

# Prediction
print('Prediction:')
import numpy as np

# Define the logistic function with the fitted parameters
def logistic(x, L, k, x0):
    return L / (1 + np.exp(-k * (x - x0)))

# Parameters from the fitted logistic model
L = 47
k = 1.0
x0 = 450

# Future scenarios: number of computers
future_computers = np.array([500, 600, 700, 800])

# Predict performance for these future scenarios
predicted_performance = logistic(future_computers, L, k, x0)

# Print results
for computers, performance in zip(future_computers, predicted_performance):
    print(f"With {computers} computers, the predicted student performance is approximately {performance:.2f}.")
```

### **Regression Analysis**

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Sample data: Replace with your actual data
data = pd.DataFrame({
    'number_of_computers': [100, 200, 300, 400, 500, 600, 700, 800],
    'student_performance': [10, 20, 30, 40, 45, 47, 48, 48]
})

# Extract features and target variable
X = data[['number_of_computers']]
y = data['student_performance']

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Predict and plot
y_pred = model.predict(X)

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Observed Data')
plt.plot(X, y_pred, color='red', label='Fitted Line')
plt.xlabel('Number of Computers')
plt.ylabel('Student Performance')
plt.title('Linear Regression Analysis')
plt.legend()
plt.grid(True)
plt.show()

# Output the coefficient and intercept
coefficient = model.coef_[0]
intercept = model.intercept_

print(f"Linear Regression Model Parameters:")
print(f"Coefficient: {coefficient}")
print(f"Intercept: {intercept}")

# Interpretation
print("\nInterpretation:")
print(f"The coefficient indicates how much the student performance increases for each additional computer.")
print(f"An increase of 1 computer results in an increase of {coefficient} units in student performance.\n")

# Prediction
print(f'Prediction:')
import pandas as pd
from sklearn.linear_model import LinearRegression

# Define the linear regression model with the fitted parameters
coefficient = 0.055
intercept = 11.25

# Future scenarios: number of computers
future_computers = np.array([500, 600, 700, 800]).reshape(-1, 1)

# Predict performance using the linear regression equation
model = LinearRegression()
model.coef_ = np.array([coefficient])
model.intercept_ = intercept
predicted_performance = model.predict(future_computers)

# Print results
for computers, performance in zip(future_computers.flatten(), predicted_performance):
    print(f"With {computers} computers, the predicted student performance is approximately {performance:.2f}.")
```
