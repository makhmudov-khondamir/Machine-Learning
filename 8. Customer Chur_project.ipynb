# **Determining customer churn**

A frequent challenge in business is understanding a customer's mood. If a business owner can anticipate whether a customer will not return or continue using the product or service, they can take proactive measures to retain the customer and change their decision. In this approach, we predict customer retention by analyzing the data of online store patrons.


### "Customer Churn Analysis Report based on the above picture"

The visualizations below depict various attributes of customers who have churned (left) versus those who have not. The data is analyzed using count plots and a histogram to understand the distribution of churn across different features.

**Gender and Churn:** The count plot for gender shows that both males and females have a significant number of non-churned customers. For females, there are approximately 2,500 non-churned customers compared to around 500 churned customers. For males, there are about 3,000 non-churned customers and approximately 750 churned customers. This suggests that gender does not have a strong impact on churn, as the ratio of churned to non-churned is similar across both genders.

**Complain and Churn:** The count plot for complaints indicates that customers who have not lodged complaints (0) are less likely to churn compared to those who have lodged complaints (1). Specifically, there are around 3,500 non-churned customers who did not complain, versus about 500 churned customers. For those who complained, there are roughly 1,000 non-churned customers and about 600 churned customers. This demonstrates a noticeable increase in churn among customers who have complained, indicating that customer dissatisfaction is a key factor in churn.

**Marital Status and Churn:** The count plot for marital status reveals that married customers have the highest count of non-churned customers, followed by single and divorced customers. For married customers, there are approximately 2,500 non-churned and 500 churned customers. Single customers have about 1,500 non-churned and 500 churned, while divorced customers have around 1,000 non-churned and 400 churned customers. This suggests that marital status may influence churn, with single and divorced customers being more prone to churn.

**Tenure and Churn:** The histogram for tenure shows that customers with shorter tenure (less than 10 months) are more likely to churn. The count of churned customers is highest in the 0-10 months range and decreases as tenure increases. This indicates that customers who have been with the company for a shorter period are more likely to leave, suggesting the need for targeted retention strategies for newer customers.

**Number of Devices Registered and Churn:** The count plot for the number of devices registered reveals that customers with 2 to 4 devices registered have the highest counts of non-churned customers. The churned customer count is relatively low across all categories, with a slight increase in churn observed for customers with 2, 4, and 6 registered devices. This suggests that the number of registered devices does not have a strong impact on churn.

**Cashback Amount and Churn:** The histogram for cashback amount shows that non-churned customers tend to receive a higher cashback amount, with peaks around 150 to 200 units. Churned customers are distributed more evenly across lower cashback amounts, with fewer peaks. This suggests that higher cashback amounts may be associated with lower churn rates.

```python
data = df[['Churn','Tenure','Complain','DaySinceLastOrder','CashbackAmount','MaritalStatus','Gender']]
#high correlation and some categorical (relative) columns were selected
```

```python
data=data.dropna()
```

```python
encoded = pd.get_dummies(data)
encoder=encoded.astype(int)
```

```python
x=encoder.drop('Churn', axis=1)
y=encoder['Churn']
```

```python
scaler=StandardScaler()
X=scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=0)
```

### **Decision Tree**

```python
DT_model = DecisionTreeClassifier(random_state=0)
DT_model.fit(x_train, y_train)
```

```python
prediction = DT_model.predict(x_test)
```

```python
print(classification_report(y_test, prediction))
```

```python
conf_mat = metrics.confusion_matrix(y_test, prediction)
sns.heatmap(conf_mat, annot=True,fmt="g")
plt.show()
```
